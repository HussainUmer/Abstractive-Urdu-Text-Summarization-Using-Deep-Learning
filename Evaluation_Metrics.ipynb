```python
# Install required libraries
# !pip install pandas sentence-transformers openpyxl

import pandas as pd
from sentence_transformers import SentenceTransformer, util
from google.colab import files


# Load the Excel file
df = pd.read_excel(file_path)

# Ensure the columns exist in your data
required_columns = ["Original Text", "Reference Summary", "Generated Summary"]
if not all(col in df.columns for col in required_columns):
    raise ValueError(f"The Excel file must have columns: {', '.join(required_columns)}")

# Load a pre-trained model for Urdu/multilingual text
# You can use "xlm-roberta-base" or "bert-base-multilingual-cased"
model = SentenceTransformer('xlm-roberta-base')

# Calculate STS (Cosine Similarity) for each row
similarity_scores = []
for _, row in df.iterrows():
    reference = row["Reference Summary"]
    generated = row["Generated Summary"]

    # Generate sentence embeddings
    ref_embedding = model.encode(reference, convert_to_tensor=True)
    gen_embedding = model.encode(generated, convert_to_tensor=True)

    # Calculate cosine similarity
    similarity = util.cos_sim(ref_embedding, gen_embedding).item()
    similarity_scores.append(similarity)

# Add STS scores to the DataFrame
df["STS_Score"] = similarity_scores

# Save results to a new Excel file
output_path = 'sts_evaluation_results.xlsx'
df.to_excel(output_path, index=False)

print(f"STS evaluation completed. Results saved to {output_path}")

# Download the file to your local machine
files.download(output_path)
```

```python
import pandas as pd
import nltk
from nltk.translate.meteor_score import meteor_score

# Download necessary NLTK resources
try:
    nltk.download('wordnet')
    nltk.download('punkt')
except:
    print("Failed to download resources. Using fallback tokenization.")


# Ensure the necessary columns exist
required_columns = ["Reference Summary", "Generated Summary"]
if not all(col in df.columns for col in required_columns):
    raise ValueError(f"The Excel file must have columns: {', '.join(required_columns)}")

# Calculate METEOR scores
meteor_scores = []
for _, row in df.iterrows():
    reference = row["Reference Summary"]
    generated = row["Generated Summary"]

    # Tokenize the summaries (with fallback)
    try:
        reference_tokens = nltk.word_tokenize(reference)
        generated_tokens = nltk.word_tokenize(generated)
    except:
        # Fallback to basic space-based tokenization
        reference_tokens = reference.split()
        generated_tokens = generated.split()

    # Calculate METEOR score
    score = meteor_score([reference_tokens], generated_tokens)
    meteor_scores.append(score)

# Add METEOR scores to the DataFrame
df["METEOR_Score"] = meteor_scores

# Save results to a new Excel file
output_path = 'meteor_evaluation_results.xlsx'
df.to_excel(output_path, index=False)

print(f"METEOR evaluation completed. Results saved to {output_path}.")
files.download(output_path)
```

```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
from google.colab import drive, files



# Ensure the necessary columns exist
required_columns = ["Reference Summary", "Generated Summary"]
if not all(col in df.columns for col in required_columns):
    raise ValueError(f"The Excel file must have columns: {', '.join(required_columns)}")

# Load BLEURT model and tokenizer from Hugging Face
model_name = "Elron/bleurt-large-512"  # Alternative BLEURT model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Calculate BLEURT scores
bleurt_scores = []
for _, row in df.iterrows():
    reference = row["Reference Summary"]
    generated = row["Generated Summary"]

    # Tokenize input
    inputs = tokenizer(reference, generated, return_tensors="pt", truncation=True, padding=True, max_length=512)

    # Compute BLEURT score
    with torch.no_grad():
        outputs = model(**inputs)
        score = outputs.logits.squeeze().item()  # Extract the score
    bleurt_scores.append(score)

# Add BLEURT scores to the DataFrame
df["BLEURT_Score"] = bleurt_scores

# Save results to an Excel file
output_path = '/content/bleurt_evaluation_results.xlsx'
df.to_excel(output_path, index=False)

print(f"BLEURT evaluation completed. Results saved to {output_path}")

# Download the file
files.download(output_path)
```
